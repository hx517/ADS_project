```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import geopandas as gpd
%pylab inline
```

```python
test_area = gpd.read_file("bin/treatment.shp")
control_area = gpd.read_file("bin/control.shp")
```

```python
test_area.columns
```

```python
fig = plt.figure(figsize=[8,10])
ax = fig.add_subplot(111)
ax.set_aspect('equal')
test_area.plot(column=u'LocationID',ax=ax,alpha=0.6)
```

```python
fig = plt.figure(figsize=[6,5])
ax = fig.add_subplot(111)
ax.set_aspect('equal')
control_area.plot(column=u'LocationID',ax=ax,alpha=1)
```
