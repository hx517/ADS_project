```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import geopandas as gpd
%pylab inline
```

```python
test_area = gpd.read_file("bin/test_area.shp")
controlgroup = gpd.read_file("bin/control_group.shp")
```

```python
test_area.columns
```

```python
fig = plt.figure(figsize=[10,13])
ax = fig.add_subplot(111)
ax.set_aspect('equal')
test_area.plot(column='Id',ax=ax,alpha=0.6)
```

```python
fig = plt.figure(figsize=[10,13])
ax = fig.add_subplot(111)
ax.set_aspect('equal')
controlgroup.plot(column='Id',ax=ax,alpha=1)
```
